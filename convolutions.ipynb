2D Convolutions
--------------------------

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread("/home/nishanth/cat.14.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img,cmap='gray')
```

```python
conv_filter = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
new_img = signal.convolve2d(img, conv_filter, boundary='symm', mode='valid')
plt.imshow(new_img,cmap='gray')
```

```python
conv_filter2 = np.ones((6,6))/img.max()
new_img = signal.convolve2d(img, conv_filter2, boundary='symm', mode='valid')
plt.imshow(new_img,cmap='gray')
```

```python
conv_filter3 = np.array([[-3,0,3],[-3,0,3],[-3,0,3]])
new_img = signal.convolve2d(img, conv_filter3, boundary='symm', mode='valid')
plt.imshow(np.absolute(new_img),cmap='gray')
```
